# Laboratorio I: Introducción a Taller de Instrumentación

Para el desarrollo de este laboratorio se planteó el siguiente orden de actividades:

-Primero, haciendo uso de las herramientas encontradas en el programa LabView se logran generar los tonos que van desde 50 hasta 5 kHz.  
-Una vez obtenidos los audios se procede a realizar las grabaciones. Estas se hacen en el mismo espacio y con el mismo parlante, para no tener variaciones entre las pruebas que se hagan.  
-Se realizan las grabaciones con los micrófonos de un teléfono celular y de una computadora. Estas grabaciones se guardan en formato wav el cual es ampliamente utilizado para almacenar grabaciones de audio de alta calidad.  
-Seguidamente se elabora una estimación del valor del SNR tomando como referencia un tono de 1 kHz.

Al iniciar el diseño, primeramente se debe de configurar el código para que este reconozca la tarjeta de sonido de la computadora 
y así poder emitir el sonido tal y como se observa en la imagen.  


Seguido de esto se configura la señal de onda que se va a generar dándole el rango de frecuencias, amplitud y demás; esto para generar la señal y graficarla. Esta señal al ir en un ciclo, se va generando punto por punto, por lo que para tener una construcción completa de la grafica y así poder visualizarla de una mejor manera, se indexa y se ingresa todo en un arreglo para luego graficar los elementos de este arreglo y a partir de acá graficar la señal con la que se construyó el sonido.  

Al finalizar esta parte, tal y como se observa en la imagen a la salida de la generación del sonido, tenemos un reset, ya que 
en LabVIEW las tareas siempre se deben de limpiar o finalizar esto para evitar el consumo de recursos o para evitar la adquisición 
de datos de otro canal el cual nos puede afectar las señales ya que nos podría causar ruido.
